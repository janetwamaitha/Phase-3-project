# Tanzania Water Wells

## Business Understanding

Tanzania is a developing country that struggles to provide clean water to its population of over 57,000,000. There are many water points already established in the country but some are in need of repair while others have failed altogether.

Our client the Government of Tanzania is looking to find patterns in wells to influence how new wells are built. The Government's goal is to provide clean and accessible water to the people of Tanzania. They currently want to expand their water network based on the existing one. This will enable them to identify what is working well and how to make improvements and for what is not working well the best way to deal with it. It could be changing the source of water or something simple like replacing the clogged pipes from salty water.

Objectives:
1. Identify what factors are contributing to the condition of the waterpoints.
2. Identify the characteristics of the functional and non-functional wells that will influence how the new wells will be built.


## Data Understanding

The data concerning the conditions of the waterpoints was recorded between 2002 and 2013. We can use 2013 as its the newest year to calculate the age of the waterpoints. Once we have the age column we can classify the waterpoints as new, old and oldest. The age of the waterpoints is most likely to influence the condition of the waterpoints.

```python
#Creating function to calculate age
def age (df):
    df['age'] = df['construction_year'].apply(lambda x: (2013 - x) if x != 0 else 0 )
    
    return df['age']
```

## Baseline Model

We will start by a single Decision Tree model. We will use the default parameters apart from the random_state that we will set to 42 for reproducability.

```python
#Defining our target variable
y = y_train['status_group']
y
```

```python
#Splitting our data
X_train, X_test, y_train, y_test = train_test_split(X_train_df, y, test_size = 0.2, random_state=42)
```

```python
#Train a Decision Tree classifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train,y_train)
```

```python
#Checking  the dimensions of split data
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
# Make predictions for test data
y_test_pred = dt.predict(X_test)

#Calculating accuracy
accuracy_score(y_test, y_test_pred)
```

The first model, simple decision tree model has an accuracy of 74% meaning that 74 out of 100 predictions are correct.  This is a realistic accuracy. We will try out a more complex model to check if the accuracy score improves. We will use a Random Forest Classifier with default parameters first and a random state of 42.

```python
#Create a function to check which features are important for each model
def feature_importances(model):
    n_features = X_train.shape[1]
    plt.figure(figsize=(8,8))
    plt.barh(range(n_features), model.feature_importances_, align='center') 
    plt.yticks(np.arange(n_features), X_train.columns.values) 
    plt.xlabel('Feature importance')
    plt.ylabel('Feature')

feature_importances(dt)
```

 The top 5 most important features for this model are: gps_height, quantity, waterpoint_type_group, funder and the installer in that order. 
 We can try tuning our parameters for the baseline model to see if the accuracy will improve. We will use entropy for our criterion and set a max_depth of 5.

```python
# Train a Decision Tree Classifier
dt_5 = DecisionTreeClassifier(criterion='entropy', max_depth=5)
dt_5.fit(X_train,y_train)
```

```python
# make predicitions on test data
y_test_pred = dt_5.predict(X_test)

#Calculating accuracy
accuracy_score(y_test, y_test_pred)
```

Tuning our parameters has reduced the score of the model from 74% to 71% so it may be best to use our baseline model with default parameters.

```python
#Train a Random Forest Classifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train,y_train)
#Parameters used
rf.get_params()
```

```python
#Making predictions on test data
y_test_pred = rf.predict(X_test)

#Calculating accuracy
accuracy_score(y_test, y_test_pred)
```

```python
#Checking important features in the Random Forest Classfier.
feature_importances(rf)
```

There's a difference in the important features in this model in that they have a higher importance value compared to the decision tree. The order is more or less the same ,gps_height is still in the lead, followed by quantity, funder, installer then waterpoint_type_group.

```python
plot_confusion_matrix(rf, X_test, y_test,
                     cmap=plt.cm.Blues)
plt.show()
```

```python
#Classification report
target_names= ['Functional', 'Functional needs Repair', 'Non-Functional']
print(classification_report(y_test, y_test_pred, target_names=target_names))
```

At first try the Random Forest Classifier has a better accuracy of 77% compared to the Decision Tree model we will try to improve this by tuning some hyperparameters. The parameters we will focus on will be:

 - n_estimators:number of trees in the forest
 - criterion: the default we've used is the 'gini', we will test with 'entropy' as well.
 - max_depth: maximum number of features considered for splitting node
 - min_samples_split: minimum number of data points placed in a node before the node is split
 - min_samples_leaf: mninimum number of data points allowed in a leaf node

We will also include a crossvalidation of 3 to reduce the problem of overfitting.

```python
#Creating  a dictionary of parameters for our gridsearch
rf_param_grid = {
    'n_estimators': [30, 150,],
    'criterion': ['gini', 'entropy'],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 3, 6],
}
```

```python
#Training the New RF classifier
rf_grid_search = GridSearchCV(rf, rf_param_grid, cv=3)
rf_grid_search.fit(X_train, y_train)
```

```python
#Making predictions on test data
y_pred_grid = rf_grid_search.predict(X_test)
#Calculating accuracy
accuracy_score(y_test, y_pred_grid)
```

```python
rf_grid_search.best_params_
```

```python
plot_confusion_matrix(rf_grid_search, X_test, y_test,
                     cmap=plt.cm.Blues)
plt.show()
```

```python
#Getting a classification report
print(classification_report(y_test, y_pred_grid, target_names=target_names))
```
